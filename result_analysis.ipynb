```python
import pandas as pd
```

```python
#the contents of all results were obtained by merging resul
df = pd.read_csv("all_results.csv")
```

```python
df.head()
```

```python
df_short = df.loc[df['dataset']=='shortterm']
df_long = df.loc[df['dataset']=='longterm']
```

```python
group_short = df_short.groupby(['dataset','algorithm', 'quantity'])
group_long = df_long.groupby(['dataset','algorithm', 'quantity'])
```

```python
df_short.loc[group_short.rmse.idxmin()].reset_index(drop=True)[['algorithm', 'quantity', 'train_time', 'rmse']]
```

```python
df_long.loc[group_long.rmse.idxmin()].reset_index(drop=True)[['algorithm', 'quantity', 'train_time', 'rmse']].sort_values(by=['quantity'])
```
